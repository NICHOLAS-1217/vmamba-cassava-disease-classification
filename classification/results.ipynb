```python
import re
import matplotlib.pyplot as plt
```

```python
validation_accuracies = []
validation_losses = []
train_losses = []

with open("./out/vssm1_tiny_0230s/20250507200359/log_rank0.txt", "r") as f:
   for line in f:
      # val acc
      if "main.py 243" in line:
         match = re.search(r'Accuracy of the network on the 3424 test images: ([\d.]+)%', line)
         if match:
               validation_accuracies.append(float(match.group(1)))
      # val loss
      if "210/214" in line:
         match = re.search(r'Loss [\d.]+ \(([\d.]+)\)', line)
         if match:
               validation_losses.append(float(match.group(1)))
      # train loss
      if "850/855" in line:
         match = re.search(r'loss [\d.]+ \(([\d.]+)\)', line)
         if match:
               train_losses.append(float(match.group(1)))

print(validation_accuracies)
print(validation_losses)
print(train_losses)
```

```python
# Accuracy Plot
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(validation_accuracies) + 1), validation_accuracies, color='orange')
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Validation Accuracy per Epoch")
plt.xticks(range(0, len(validation_accuracies) + 1, 10))  # every 10 epochs
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Loss Plot (Train & Validation)
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(train_losses) + 1), train_losses, color='blue', label='Train Loss')
plt.plot(range(1, len(validation_losses) + 1), validation_losses, color='orange', label='Val Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training & Validation Loss per Epoch")
plt.ylim(0, 2)
plt.xticks(range(0, max(len(train_losses), len(validation_losses)) + 1, 10))  # every 10 epochs
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
